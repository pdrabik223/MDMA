```python


with open("AI3MSPRO_triangle_3d.gcode", "r") as file:
    gcode_file = np.array(file.readlines())


print(gcode_file)
print(len(gcode_file))

# # if on prusa slicer:
# layer_change_lines = np.where(gcode_file == ";LAYER_CHANGE\n")[0]

layers = [[]]
for line in gcode_file:
    # cura
    if ';LAYER' in line:
        layers.append([])
    # prusa slicer
    # if 'LAYER_CHANGE' in line:
    #   layers.append([])
    
    layers[-1].append(line)
    
print(layers)
print("no_layers: ", len(layers))
```

```python

current_height = 0.0

# printer._line_counter = 0

# printer.send_and_await("G28")
# printer.send_and_await("G1 Z30")
# for i in range(10):
#     printer.send_and_await("G1 X100 Y100")
#     printer.send_and_await("G1 X10 Y100")
#     printer.send_and_await("G1 X10 Y10")
#     printer.send_and_await("G1 X100 Y10")

for layer in layers:
    for line in layer:
        gcode = line.split(';')[0]
        gcode = gcode.strip()
        if len(gcode) > 0:
            printer.send_and_await(gcode)
            


    print("layer done") 
    # scan now!
    current_height = 0.0
```
